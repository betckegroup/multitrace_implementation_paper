```python
import bempp.api 

def function(x, y, i, j):
    zx = 0
    zy = 0
    if i == 0:
        zx = np.sin(x * np.pi * 2)
    if i == 1:
        if x <= 0.25:
            zx = np.sin(x * np.pi * 4)
    if i == 2:
        if x > 0.25:
            zx = -np.sin(x * np.pi * 4)
    if i == 3:
        if x <= 0.5 / 3.0:
            zx = np.sin(x * np.pi * 6)
    if i == 4:
        if x > 0.5 / 3.0 and x <= 1 / 3.0:
            zx = -np.sin(x * np.pi * 6)
    if i == 5:
        if x > 1 / 3.0:
            zx = np.sin(x * np.pi * 6)

    if j == 0:
        zy = np.sin(y * np.pi * 2)

    if j == 1:
        if y <= 0.25:
            zy = np.sin(y * np.pi * 4)

    if j == 2:
        if y > 0.25:
            zy = -np.sin(y * np.pi * 4)

    if j == 3:
        if y <= 0.5 / 3.0:
            zy = np.sin(y * np.pi * 6)

    if j == 4:
        if y > 0.5 / 3.0 and y <= 1 / 3.0:
            zy = -np.sin(y * np.pi * 6)
    if j == 5:
        if y > 1 / 3.0:
            zy = np.sin(y * np.pi * 6)

    return zx * zy


import numpy as np
N = 100
x = np.linspace(0, .5, N)
y = np.linspace(0, .5, N)


```

```python
xv, yv = np.meshgrid(x, y)

from matplotlib import pyplot as plt
```

```python
zv = function(xv,yv,0,0)
```

```python
plt.pcolor(xv,yv,zv)
plt.show()
```
